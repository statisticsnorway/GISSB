# 2021 vs. 2022

```r
resultater_100_tilfeldige_adresser_2021 <- read_SSB("ssb-prod-dapla-felles-data-delt/GIS/Vegnett_benchmark/Resultater_Sindre/2021_resultater_tilfeldige_adresser_100.parquet") %>%
dplyr::rename(minutes_2021 = minutes)

resultater_100_tilfeldige_adresser_2022 <- read_SSB("ssb-prod-dapla-felles-data-delt/GIS/Vegnett_benchmark/Resultater_Sindre/2022_resultater_tilfeldige_adresser_100.parquet") %>%
dplyr::rename(minutes_2022 = minutes)

resultater_100_tilfeldige_adresser <- left_join(resultater_100_tilfeldige_adresser_2021, resultater_100_tilfeldige_adresser_2022, by = c("idx", "idx_to")) %>%
dplyr::mutate(diff = minutes_2021-minutes_2022)

# head(resultater_100_tilfeldige_adresser_2021)
# head(resultater_100_tilfeldige_adresser_2022)

resultater_100_tilfeldige_adresser %>%
dplyr::arrange(desc(diff))
```
